```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
import joblib

#Oszlopok tiszt√≠t√°sa (opcion√°lis)
df = df.drop(columns=["Time", "V10", "V14", "V17"], errors="ignore")

#C√©l √©s jellemz≈ëk sz√©tv√°laszt√°sa
X = df[["V1", "V2", "V3", "V4"]].copy()
X["Amount"] = df["Amount"]
y = df["Class"]

#Normaliz√°l√°s
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Train-test feloszt√°s (80-20)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y, random_state=42)

#Modell l√©trehoz√°sa √©s tan√≠t√°sa
logreg_model = LogisticRegression(max_iter=1000)
logreg_model.fit(X_train, y_train)

#El≈ërejelz√©s
y_pred = logreg_model.predict(X_test)

#Ki√©rt√©kel√©s
print("üîç Pontoss√°g (Accuracy):", accuracy_score(y_test, y_pred))
print("\nüìä Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nüìà R√©szletes jelent√©s:\n", classification_report(y_test, y_pred))

#Modell ment√©se apphoz
joblib.dump(logreg_model, "logreg_model.pkl")

```

```python
#K√∂nyvt√°rak import√°l√°sa
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Adatok bet√∂lt√©se (felt√©telezz√ºk, hogy df m√°r l√©tezik, pl. df = pd.read_csv("..."))
df = df.drop(columns=["Time", "V10", "V14", "V17"], errors="ignore")  # felesleges attrib√∫tumok eldob√°sa

# C√©lv√°ltoz√≥ √©s input sz√©tv√°laszt√°sa
X = df[["V1", "V2", "V3", "V4"]]
X["Amount"] = df["Amount"]
y = df["Class"]

# Standardiz√°l√°s
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Zaj hozz√°ad√°sa √©s kerek√≠t√©s
np.random.seed(42)
X_noisy = X_scaled + np.random.normal(0, 2.0, X_scaled.shape)  # er≈ës zaj
X_rounded = np.round(X_noisy, 1)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_rounded, y, test_size=0.3, stratify=y, random_state=42)

# Random Forest modell l√©trehoz√°sa √©s tan√≠t√°sa
rf_model = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=42)
rf_model.fit(X_train, y_train)
rf_y_pred = rf_model.predict(X_test)

# Eredm√©nyek √©rt√©kel√©se
rf_accuracy = accuracy_score(y_test, rf_y_pred)
rf_cm = confusion_matrix(y_test, rf_y_pred)
rf_report = classification_report(y_test, rf_y_pred)

print(" Random Forest Eredm√©nyek:")
print(f" Pontoss√°g (Accuracy): {rf_accuracy}")
print("\n Confusion Matrix:\n", rf_cm)
print("\nR√©szletes jelent√©s:\n", rf_report)
joblib.dump(rf_model, "random_forest_model.pkl")
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np

# Adat zajos√≠t√°sa
X_noisy = X_scaled + np.random.normal(0, 2.0, X_scaled.shape)
X_rounded = np.round(X_noisy, 1)

# Egyszer≈±, lebut√≠tott XGBoost modell be√°ll√≠t√°sa
xgb_model = XGBClassifier(
    n_estimators=5,           # Minim√°lis boosting l√©p√©sek
    max_depth=1,              # Csak d√∂nt√©si r√∂nk√∂k (gyenge tanul√≥k)
    learning_rate=0.05,       # Lass√∫ tanul√°s, hogy ne tanuljon t√∫l gyorsan
    subsample=0.3,            # Csak 30% adat minden iter√°ci√≥ban
    colsample_bytree=0.3,     # Csak 30% v√°ltoz√≥ minden iter√°ci√≥ban
    use_label_encoder=False,
    eval_metric='logloss',
    verbosity=0,
    random_state=42
)

# Modell tan√≠t√°sa
xgb_model.fit(X_train, y_train)

# El≈ërejelz√©s teszt adaton
y_pred_xgb = xgb_model.predict(X_test)

# Ki√©rt√©kel√©s
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
conf_matrix_xgb = confusion_matrix(y_test, y_pred_xgb)
report_xgb = classification_report(y_test, y_pred_xgb)

# Eredm√©nyek megjelen√≠t√©se
print("XGBoost Eredm√©nyek:")
print(f"Pontoss√°g (Accuracy): {accuracy_xgb}")
print("\nConfusion Matrix:\n", conf_matrix_xgb)
print("\nR√©szletes jelent√©s:\n", report_xgb)


# Modell tan√≠t√°sa
xgb_model.fit(X_train, y_train)

# El≈ërejelz√©s teszt adaton
y_pred_xgb = xgb_model.predict(X_test)

# Ki√©rt√©kel√©s
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
conf_matrix_xgb = confusion_matrix(y_test, y_pred_xgb)
report_xgb = classification_report(y_test, y_pred_xgb)

# Eredm√©nyek megjelen√≠t√©se
print("XGBoost Eredm√©nyek:")
print(f"Pontoss√°g (Accuracy): {accuracy_xgb}")
print("\nConfusion Matrix:\n", conf_matrix_xgb)
print("\nR√©szletes jelent√©s:\n", report_xgb)
joblib.dump(xgb_model, "xgb_model.pkl")
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np

# Adat zajos√≠t√°sa
X_noisy = X_scaled + np.random.normal(0, 2.0, X_scaled.shape)
X_rounded = np.round(X_noisy, 1)

# Gyenge (lebut√≠tott) MLP modell be√°ll√≠t√°sa
mlp = MLPClassifier(
    hidden_layer_sizes=(10,),     # Egy rejtett r√©teg, mind√∂ssze 10 neuronnal
    activation='tanh',            # Egyszer≈±bb aktiv√°ci√≥s f√ºggv√©ny
    solver='sgd',                 # Sztochasztikus gradient descent (lassabb tanul√°s)
    learning_rate_init=0.01,      # Alacsony kezd≈ë tanul√°si r√°ta
    max_iter=100,                 # Kev√©s iter√°ci√≥
    random_state=42
)

# Tan√≠t√°s
mlp.fit(X_train, y_train)

# El≈ërejelz√©s
y_pred_mlp = mlp.predict(X_test)

# Ki√©rt√©kel√©s
acc_mlp = accuracy_score(y_test, y_pred_mlp)
cm_mlp = confusion_matrix(y_test, y_pred_mlp)
report_mlp = classification_report(y_test, y_pred_mlp)

# Eredm√©nyek megjelen√≠t√©se
print("Neur√°lis H√°l√≥ (MLP) Eredm√©nyek:")
print(f"Pontoss√°g (Accuracy): {acc_mlp}")
print("\nConfusion Matrix:\n", cm_mlp)
print("\nR√©szletes jelent√©s:\n", report_mlp)
joblib.dump(mlp, 'mlp_model.pkl')
```
